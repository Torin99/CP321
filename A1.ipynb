## Question 1

Compare the mean prices of Regular Unleaded Gasoline among different cities ('Ottawa', 'Toronto West/Ouest', 'Toronto East/Est', 'Windsor','London', 'Peterborough', 'Sudbury', 'Sault Saint Marie', 'Thunder Bay', 'North Bay', 'Timmins', 'Kenora', 'Parry Sound',"St. Catharine's") in 2023, and plot a corresponding bar chart. Please ensure that the bars are sorted by the price amounts.

```python
# Filter based on Fuel Type and Year
df_reg_2023 = df[(df['Fuel Type'] == 'Regular Unleaded Gasoline') & (pd.to_datetime(df['Date']).dt.year == 2023)].copy()

# Select relevant city columns
city_columns = df_reg_2023.columns[2:16]

# Calculate average price
df_reg_2023['Average Price'] = df_reg_2023[city_columns].mean(axis=1)

# Calculate average prices for each city
average_prices = df_reg_2023.mean(axis=0, numeric_only=True)[1:15].sort_values()
```

```python
plt.figure(figsize=(10, 6))
average_prices.plot(kind='bar')
plt.xlabel('City')
plt.ylabel('Average Price')
plt.title('Average Gas Price for Ontario Cities in 2023')
plt.xticks(rotation=45, ha='right')
plt.show()
```

## Question 2
Plot the corresponding dot plot for the same information presented in Question 1.

```python
plt.figure(figsize=(10, 6))
plt.scatter(average_prices.index, average_prices, s=100)
plt.xlabel('City')
plt.ylabel('Average Price')
plt.title('Average Gas Price for Ontario Cities in 2023')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility
plt.show()
```

## Question 3

Plot a histogram of the Regular Unleaded Gasoline prices in Toronto West/Ouest for 2023, using 15 bins.

```python
toronto_data = df_reg_2023['Toronto West/Ouest']

# Plot histogram
plt.hist(toronto_data, bins=15, color='blue', edgecolor='black')

# Add title and labels
plt.title('Regular Unleaded Gas Prices - Toronto West, 2023')
plt.xlabel('Average Price')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

## Question 4

Plot the normalized ECDF (Empirical Cumulative Distribution Function) of the Regular Unleaded Gasoline prices in Toronto West/Ouest for 2023.

```python
sorted_toronto_data = np.sort(toronto_data)

# Create an array of equally spaced values to represent the cumulative probabilities
cumulative_prob = np.arange(1, len(sorted_toronto_data) + 1) / len(sorted_toronto_data)

# Plot the ECDF
plt.plot(sorted_toronto_data, cumulative_prob, marker='.', linestyle='none')

# Add title and labels
plt.title('ECDF of Regular Unleaded Gas Prices - Toronto West, 2023')
plt.xlabel('Average Price')
plt.ylabel('Cumulative Probability')

# Show the plot
plt.show()
```

## Question 5

Plot a boxplot to display the distribution of Regular Unleaded Gasoline prices across different cities ('Ottawa', 'Toronto West/Ouest', 'Toronto East/Est', 'Windsor','London', 'Peterborough', 'Sudbury', 'Sault Saint Marie', 'Thunder Bay', 'North Bay', 'Timmins', 'Kenora', 'Parry Sound',"St. Catharine's") from 2021 onwards (including 2021).


```python
# Filter based on Fuel Type and Year
df_reg_2021up = df[(df['Fuel Type'] == 'Regular Unleaded Gasoline') & (pd.to_datetime(df['Date']).dt.year >= 2021)].copy()
city_columns = df_reg_2021up.columns[2:16]

df_cities_2021up = df_reg_2021up[city_columns]
```

```python
# Plot boxplot
plt.figure(figsize=(12, 8))
sns.boxplot(data=df_cities_2021up)
plt.title('Regular Unleaded Gasoline Prices Across Cities (2021 and onwards)')
plt.xlabel('City')
plt.ylabel('Gas Price')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Show the plot
plt.show()
```
